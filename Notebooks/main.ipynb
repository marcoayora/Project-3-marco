```python
def normalize_to_10(x):
    min_val = x.min()
    max_val = x.max()
    return round(10 * (x - min_val) / (max_val - min_val), 2)

#  Select only integer columns for normalization
cols_to_normalize = df.select_dtypes(include='int64').columns[0:]

normalized_offices = df.copy()

# Apply the normalization function to selected columns
normalized_offices[cols_to_normalize] = normalized_offices[cols_to_normalize].apply(normalize_to_10)

# Print the dataframe to see the result
normalized_offices
```
